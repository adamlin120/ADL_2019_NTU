```python
log_file = '../elmo/models/report_1/log.txt'
```

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt

with open(log_file) as f:
    hist = f.readlines()
hist = list(map(lambda x: x.strip().split(','), hist))
loss = {'train': [], 'val': []}
steps = []
for line in hist:
    if line[0].startswith('epoch'):
        steps.append(int(line[1].split(':')[-1]))
    else:
        loss[line[0]].append(float(line[2]))

num_steps = min(len(steps), len(loss['train']), len(loss['val']))
steps = steps[:num_steps]
loss['train'] = loss['train'][:num_steps]
loss['val'] = loss['val'][:num_steps]

steps = np.array(steps)
loss = {'train': np.array(loss['train']), 'val': np.array(loss['val'])}
perplexity = {'train': np.exp(loss['train']), 'val': np.exp(loss['val'])}

main = os.path.basename(os.path.dirname(os.path.normpath(log_file)))
plt.title(main)
plt.xlabel('Step')
plt.ylabel('Perplexity')
plt.plot(steps, perplexity['train'], color='grey', linewidth=2, label="Train")
plt.plot(steps, perplexity['val'], color='orange', linewidth=2, label="Dev")
plt.legend()
plt.savefig(f'{main}_perplexity.png')
```
